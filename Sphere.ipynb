```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
radius = 1

# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(0, 1.25, 50)
t = np.linspace(0, 2*np.pi, 360)
p = np.linspace(-np.pi/2, np.pi/2, 180)

T, P = np.meshgrid(t, p)

# Express the mesh in the cartesian system.
X = radius*np.cos(T)*np.cos(P)
Y = radius*np.sin(T)*np.cos(P)
Z = radius*np.sin(P)

# Plot the surface.
ax.plot_surface(X, Y, Z, alpha = 1, cmap = 'winter')

# Location of circle in t, p
tc = np.pi/2
pc = np.pi/4

# Adding to list
XC = [0, 1.5*radius*np.cos(tc)*np.cos(pc)]
YC = [0, 1.5*radius*np.sin(tc)*np.cos(pc)]
ZC = [0, 1.5*radius*np.sin(pc)]
ax.plot(XC, YC, ZC, 'k')

# Tweak the limits and add labels.
ax.set_xlim(-radius, radius)
ax.set_ylim(-radius, radius)
ax.set_zlim(-radius, radius)

ax.set_xlabel(r'X')
ax.set_ylabel(r'Y')
ax.set_zlabel(r'Z')
```
