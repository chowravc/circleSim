```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import capCreator as cC
import time as t
import numpy as np
%matplotlib notebook
```

```python
#choose radius (of sphere), theta of cap and euler angles
theta = np.pi/4
radius = 1
angles = [0.2, 1.1, 3]
resolution = 0
```

```python
now = t.time()
xf, yf, zf = cC.cap(theta, radius, angles, resolution)
print("Time to create island (seconds):",str(t.time() - now))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(-1.5, 1.5)

ax.plot_surface(xf, yf, zf, rstride = 4, cstride = 4, alpha = 0.7, color = 'r')

plt.show()
```
