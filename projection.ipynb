```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import time as t
import numpy as np

import capCreator as cC
import hammerProjection as hP

%matplotlib notebook
```

```python
#choose radius (of sphere), theta of cap and euler angles
theta = np.pi/4
radius = 1
angles = [0, 0, 5]
resolution = 100
```

```python
#now = t.time()
angles = [np.pi/2, 0, 0]
xf, yf, zf = cC.cap(theta, radius, angles, resolution)
#print("Time to create island (seconds):",str(t.time() - now))
```

```python
xh, yh = hP.project(xf, yf, zf, radius)
```

```python
plt.xlim(-2, 2)
plt.ylim(-2, 2)

plt.plot(xh, yh)
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(-1.5, 1.5)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')


ax.plot_surface(xf, yf, zf, rstride = 4, cstride = 4, alpha = 0.7, color = 'r')

plt.show()
```
